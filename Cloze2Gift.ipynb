```python
def reformatQuestion(reQuestion):
    return reQuestion.replace("~","\n~").replace("=","\n=")
    
def giftQuestion(name,question):
    m=re.match(ur"^(.*)\{\d+:MC:(.*)\}(.*)",question)
    if m:
        giftCorps=reformatQuestion(m.group(2))
        return"::"+name+"::[html]"+m.group(1)+'{'+giftCorps+"}\n"+m.group(3)+"\n"
    else:
        print "no question found"
```
